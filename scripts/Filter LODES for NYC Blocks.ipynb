```python
import pandas as pd
import numpy as np
```

```python
lodes = pd.read_csv('../ny_od_main_JT00_2013.csv', dtype={'w_geocode': str, 'h_geocode': str})
```

```python
lodes['w_county'] = lodes['w_geocode'].str[2:5]
lodes['h_county'] = lodes['h_geocode'].str[2:5]
```

```python
lodes
```

```python
counties = '005|047|061|081|085'
```

```python
nyc_lodes = lodes[lodes['w_county'].str.contains(counties)&lodes['h_county'].str.contains(counties)]
```

```python
nyc_lodes
```

```python
nyc_lodes.to_csv('lodes_nyc.csv')
```

```python
blocks = np.unique(np.concatenate((nyc_lodes['w_geocode'], nyc_lodes['h_geocode'])))
```

```python
blocks_df = pd.DataFrame({'block': blocks})
blocks_df.to_csv('nyc_blocks_in_lodes.csv')
```
